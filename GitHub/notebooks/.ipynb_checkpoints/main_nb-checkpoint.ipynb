```python
from pathlib import Path
import pandas as pd
import pyyaml as yaml

# Core
from src.ma_migration.core.contracts import DomainContract

# Customers domain
from src.ma_migration.domains.customers import transforms as T
from src.ma_migration.domains.customers import validations as V
from src.ma_migration.domains.customers.pipeline import prepare_customers
from src.ma_migration.contracts.loader import load_contract

# Optional utils
pd.set_option("display.max_columns", 100)
```

```python
CONTRACT_PATH = f'{ROOT}/contracts/customers/customers.yaml'

raw_contract = load_contract(CONTRACT_PATH)
contract = DomainContract(raw=raw_contract, path=CONTRACT_PATH)

contract.raw  # Display the YAML content
```

```python
INPUT_PATH = Path("INPUT/Customers_MasterFile.xlsx")

df_customers = pd.read_excel(INPUT_PATH, sheet_name="Customers")  # or contract.sheet.default if defined

df_customers.head()
df_customers.shape
```

```python
missing = []
for key, meta in contract.raw["columns"].items():
    colname = meta["name"]
    if meta.get("required") and colname not in df_customers.columns:
        missing.append(colname)

if missing:
    print("❌ Missing required columns:")
    for col in missing:
        print("   -", col)
```

```python
result = prepare_customers(
    df_customers,
    contract=contract,
    strict=False   # exploration mode — recommended while building
)

result
```
